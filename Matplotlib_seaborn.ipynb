```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
y1 = [5,4,7,8,9,12]

gender = ['female','male','female','male','male','female']

plt.figure(figsize=(5,3))
plt.plot(x,y,color='blue',marker='*',linestyle = 'dashed',label = "Product A")
plt.grid(True)
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line plot")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y,color='blue',marker='o',label = "Product A")
plt.grid(True)
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(x,color='purple',bins = 12,label = "Product A")
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Histogram plot")
plt.show()
```

```python
salary = [15000,23000,70000,50000]
label = ['Intern','Employee','Senior','Manager']

plt.pie(salary,labels=label,colors=['red','blue','purple','green'],autopct='%2.2f%%')
plt.legend()
plt.title("Pie chart")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color='purple',label = "Product A")
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar plot")
plt.show()
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df
```

```python
sns.lineplot(x = 'size',y='total_bill',data = tips_df,hue='sex',style="day",palette = "magma")
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],hist = True,color = 'r',bins = 100)
plt.show()
```

```python
sns.barplot(x = 'day',y = 'total_bill',data = tips_df,hue = 'sex',hue_order=['Female','Male'])
plt.show()
```

```python
sns.scatterplot(x = 'total_bill',y='tip',data = tips_df,hue = "sex")
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x= 'total_bill',y = 'tip',data = tips_df,hue = 'sex')
sns.barplot(x = 'total_bill',y = 'tip',data = tips_df)
plt.show()
```
